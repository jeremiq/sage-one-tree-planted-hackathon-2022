### Plot out disadvantaged group in CA

```python
# remove the leading 0 
# ref: https://stackoverflow.com/questions/68905161/remove-leading-zeroes-pandas#:~:text=Remove%20all%20leading%20zeros%3A,on%20the%20string%20argument%20passed).
ca_locations = usa_locations[(usa_locations.SF == "California")]
ca_locations["GEOID10"] = ca_locations["GEOID10"].str.lstrip("0")
```

```python
# Note: the `Census tract ID` has data type int, which needs to be converted to string because `GEOID10` is string data type
ca_disadvantaage_group = set(df2[(df2["State/Territory"] == "California") & (df2["Identified as disadvantaged"] == True)]["Census tract ID"].astype(str).tolist())
```

```python
ca_under = ca_locations[ca_locations["GEOID10"].isin(ca_disadvantaage_group)]
ca_under.plot(cmap="jet", column="GEOID10", figsize=(10,10))
```
